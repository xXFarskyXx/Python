```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
path = "Doiy bacteria.jpg"
bgr = cv2.imread(path)
bgr = cv2.resize(bgr, (int(bgr.shape[1]/4) , int(bgr.shape[0]/4)))

gray = cv2.cvtColor(bgr , cv2.COLOR_BGR2RGB)

plt.imshow(gray , cmap="gray")
```

```python
blur = cv2.GaussianBlur(gray , (1001 , 1001) , cv2.BORDER_DEFAULT)
canny = cv2.Canny(blur , 1 , 25)
plt.imshow(canny , cmap="gray")
```

```python
contour , hierarchy = cv2.findContours(canny , cv2.RETR_LIST , cv2.CHAIN_APPROX_NONE)
blank = np.zeros_like(bgr)
cv2.drawContours(blank , contour , -1 , (255 , 255 , 0) , thickness = 2)
plt.imshow(blank)
```
