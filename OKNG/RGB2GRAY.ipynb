```python
import cv2
import matplotlib.pyplot as plt
```

```python
egg = []

for i in range(6):
    path = "Egg_dataset\Egg" + str(i + 1) + ".jpg"
    img = cv2.imread(path)
    egg.append(img)

    plt.subplot(3 , 3 , i+1)
    plt.imshow(egg[i])
```

```python
egg_gray = []

for i in range(6):
    egg_gray.append(cv2.cvtColor(egg[i] , cv2.COLOR_BGR2GRAY))
    
    plt.subplot(3 , 3 , i+1)
    plt.imshow(egg_gray[i] , cmap = "gray")
```

```python
#Save images
for i in range(6):
    name = "./egg_gray" + str(i + 1) + ".png" 
    cv2.imwrite(name , egg_gray[i])
```
