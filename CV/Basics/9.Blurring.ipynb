```python
import cv2
import numpy
from matplotlib import pyplot as plt
```

```python
path = "..\opencv-course-master\Resources\Photos\cats.jpg"
img = cv2.imread(path)
rgb = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)

plt.imshow(rgb)
```

```python
#Average Blur
average = cv2.blur(rgb , (3 , 3))

plt.imshow(average)
```

**Gaussian Blur**
$$
G(x , y) = \frac {1}{2 \pi \sigma^{2}} \cdot e^{-\frac{x^{2} + y^{2}}{2 \sigma^{2}}}
$$

where  
- G(x,y) is the value of the two-dimensional Gaussian function at position (x,y).  
- σ (sigma) is the standard deviation of the distribution, controlling the spread of the Gaussian.

```python
#Gaussian Blur
gauss = cv2.GaussianBlur(rgb , (7 , 7) , 0)

plt.imshow(gauss)
```

**Bilateral Blur**
$$
BF(I , p) = \frac{1}{W~p} \sum_{q \in{S}} G~s(||p - q||) \cdot G~r(|I(p) - I(q)|) \cdot I(q)
$$
where
- BF(I,p) is the value of the bilateral filter at pixel p in the image I.
- $W~s$ is the normalization factor, calculated as the sum of the bilateral filter weights over all pixels in the spatial domain S.
- p and q are pixel positions.
- $G~s(||p - q||)$ is the spatial Gaussian filter, dependent on the spatial distance between pixels p and q.
- $G~r(|I(p) - I(q)|)$ is the intensity range filter, dependent on the intensity difference between pixels p and q.
- $I(p) and I(q)$ are the intensity values of pixels p and q, respectively.

```python
#Bilateral Blur
bilat = cv2.bilateralFilter(rgb , 10 , 35 , 15)

plt.imshow(bilat)
```
